```python
import csv
from heap import Heap

heap_data = []

with open('birthday.csv', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        name = row['이름'].strip()
        birth = row['생년월일8자리(예.20040101)'].strip()
        if birth:
            try:
                birth_int = int(birth)
                heap_data.append((birth_int, name))
            except ValueError:
                continue

heap = Heap(heap_data)
heap.buildHeap()

# 생일이 늦은(숫자가 큰) 순서로 10명 출력
for _ in range(10):
    if not heap.isEmpty():
        birth, name = heap.deleteMax()
        print(f"{name} {birth}")




```

```python
import heapq

class KthLargest:

    def __init__(self, k: int, nums: list[int]):
        self.k = k
        self.min_heap = nums
        heapq.heapify(self.min_heap)

        # k개만 남기고 나머지는 제거
        while len(self.min_heap) > k:
            heapq.heappop(self.min_heap)

    def add(self, val: int) -> int:
        heapq.heappush(self.min_heap, val)
        # k개를 초과하면 가장 작은 값 제거
        if len(self.min_heap) > self.k:
            heapq.heappop(self.min_heap)
        return self.min_heap[0]  # k번째로 큰 값은 항상 heap의 top

kth = KthLargest(3, [4, 5, 8, 2])
print(kth.add(3))  # return 4
print(kth.add(5))  # return 5
print(kth.add(10)) # return 5
print(kth.add(9))  # return 8
print(kth.add(4))  # return 8
```
